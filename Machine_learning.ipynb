```python
#Replace all input sets fields with the following:

# yes to 1 and no to 0
#for furnished 2, semi furnished 1 and unfurnished 0

file_2 = df.replace(to_replace = ['yes','no','furnished','semi-furnished','unfurnished'], value=[1,0,2,1,0])
```

```python
file_2
```

```python
#best practice for naming output dataset (splitting datasets) X===> input datasets, y===> output datasets
```

```python
X = file_2.drop(columns=['price']) #input dataset
X
```

```python
y =file_2['price'] #output dataset
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2) # test size should be 20% as per best practice
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
check_acc = model.predict(X_test)
scores = accuracy_score(y_test, check_acc)
scores
```

```python
predict_house_price = model.predict([[8960,4,1,2,1,1,0,0,1,1,0,0]])
```
